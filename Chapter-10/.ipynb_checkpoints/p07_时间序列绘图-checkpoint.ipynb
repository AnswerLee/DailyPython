```python
# 时间序列绘图
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from dateutil.parser import parse
from pandas import DataFrame,Series

%matplotlib inline
```

```python
close_px_all = pd.read_csv('../dataset/stock_px.csv',
                           parse_dates=True,index_col=0)
close_px_all.head()
```

```python
close_px = close_px_all[['AAPL','MSFT','XOM']]
close_px.head()
```

```python
close_px.resample('B').pad().head(10) # 用工作日来采样
```

```python
close_px.shape
```

```python
close_px = close_px.resample('B').pad()
close_px.head()
```

```python
close_px['AAPL'].plot() # Series 的画图方法
```

```python
close_px.loc['2009'].shape
```

```python
close_px.loc['2009'].plot()# 定位到2009年，作图
```

```python
# 季度价格变化

close_px.resample('Q-DEC').pad().head()
```

```python
appl_q = close_px['AAPL'].resample('Q-DEC').pad()# 季度价格变化
appl_q.loc['2009':].plot()# 09 年到现在每个季度的价格变化
```
