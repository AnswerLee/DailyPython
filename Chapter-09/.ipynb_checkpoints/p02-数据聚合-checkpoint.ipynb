```python
# 数据聚合
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
```

```python
df = DataFrame({'key1':['a','a','b','b','a'],
               'key2':['one','two','one','two','one'],
               'data1':np.random.randn(5),
               'data2':np.random.randn(5)})
df
```

```python
grouped = df.groupby('key1')
grouped['data1'].quantile(0.9)# 计算分组之后的分位数
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
grouped.agg(peak_to_peak) # 对分组之后的数据使用自定义聚合函数
```

```python
grouped.describe()
```

```python
# 优化过的聚合函数：
# count：     非NA值的数量
# sum：       非NA值的和
# mean：      非NA值的平均数
# median：    非NA值的中位数
# std/var：   无偏（分母为n - 1）的标准差和方差
# min/max：   非NA值的最小/最大值
# prod：      非NA值的积
# first/last：第一个/最后一个非NA值
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 新加一列，小费与账单金额的比例
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips.head()
```

```python
# 面向列的多函数应用
```

```python
# 根据用餐时间做分组
grouped = tips.groupby(['smoker','time'])
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
tips.groupby(['smoker','time']).agg({'tip_pct':np.mean})
```

```python
grouped_pct.agg(['mean','std',peak_to_peak])
```

```python
tips.groupby(['smoker','time']).agg({'tip_pct':['mean','std',peak_to_peak]})
```

```python
#列的重命名 ，命名为foo  bar
grouped_pct.agg([('foo','mean'),('bar',np.std)])
```

```python
# 对group后的两个字段分别作用functions
functions = ['count','mean','max']
result = grouped['tip_pct','total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
ftuples = [('Durchschnitt','mean'),('Abweichung',np.var)]
grouped['tip_pct','total_bill'].agg(ftuples)
```

```python
#不同的列对应不同的函数
grouped.agg({'tip' : np.max, 'size' : 'sum'})
```

```python
# 每一列可以对应不同数量的函数
grouped.agg({'tip_pct':['mean','sum','count','std'],
            'size':['sum','min','max']})
```

```python
# 以“无索引”的形式返回聚合数据
# 把原来的索引变成列
tips.groupby(['smoker','time'],as_index=False).mean()
```
